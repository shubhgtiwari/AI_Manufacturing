```python
import pandas as pd
import os 
from sqlalchemy import create_engine
from dotenv import load_dotenv
from urllib.parse import quote_plus
```

```python
load_dotenv('.env')
```

```python
user = os.getenv("POSTGRES_USER")
password = os.getenv("POSTGRES_PASSWORD")
db = os.getenv("POSTGRES_DB")
host = "localhost"
```

```python
encoded_password = quote_plus(password)
db_str = f"postgresql://{user}:{encoded_password}@{host}:5432/{db}"
engine = create_engine(db_str)

print(f"Connecting to: {db}")
```

```python
query = "SELECT * FROM sap_sales_flat"
df = pd.read_sql(query, engine)

query = "SELECT * FROM iot_sensors"
dfiot = pd.read_sql(query,engine)
```

```python
df.info()
```

```python
keep_cols = [
    'VBELN',   
    'AUDAT',  
    'MATNR',   
    'NETWR',   
    'CATEGORY',
    'order_status' 
]
```

```python
df_clean = df[keep_cols].copy()
print(f"Trimmed Shape:  {df_clean.shape}")
```

```python
df_clean['CATEGORY'].unique()
```

```python
df_clean['order_status'].unique()
```

```python
df_clean = df_clean[df_clean['order_status'] == 'delivered']
```

```python
df_clean[df_clean['NETWR'] == 0.85]
```

```python
df_clean.loc[:, 'AUDAT'] = pd.to_datetime(df_clean['AUDAT'])
```

```python
df_clean.info()
```

```python
df_clean[df_clean['CATEGORY'].isnull() == True]
```

```python
df_clean['CATEGORY'] = df_clean['CATEGORY'].fillna('unknown')
```

```python
df_sales = df_clean.dropna()
```

```python
df_sales.info()
```

```python
df_sales.to_sql('sap_clean_sales_data', engine, if_exists='replace', index=False)
```
